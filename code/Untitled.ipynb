```python
text = ",".join(review for review in df.notas_de_cata.astype(str)).lower()
stopwords = set(STOPWORDS)
stopwords.update(["blanco","negro","secos","de","la","del","nan"])

wordcloud = WordCloud(stopwords=stopwords,width=1000, height=600,colormap='Paired').generate(str(text))


plt.figure( figsize=(40,20))
plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
mask = np.array(Image.open("cafe.png"))
```

```python
wc = WordCloud(mask=mask, 
              width=1000, max_words=200, height=1000,
              random_state=1)
wc.generate(text)
plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
```
